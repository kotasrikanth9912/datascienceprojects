```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel("C:\\Users\\Kotas\\OneDrive\\Desktop\\DATA SCIENCE\\Dataset .xlsx")

# Display dataset info
print(df.info())

# ----------------------------------------
# 1. Visualize Restaurant Locations on Map
# ----------------------------------------
plt.figure(figsize=(10, 6))
plt.scatter(df["Longitude"], df["Latitude"], alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Distribution of Restaurants")
plt.show()

# ----------------------------------------
# 2. Distribution Across Cities
# ----------------------------------------
top_cities = df["City"].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_cities.plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities with Highest Number of Restaurants")
plt.show()

# ----------------------------------------
# Distribution Across Countries
# ----------------------------------------
top_countries = df["Country Code"].value_counts()

plt.figure(figsize=(8, 8))
top_countries.plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Restaurants Across Countries")
plt.ylabel("")
plt.show()

# ----------------------------------------
# 3. Correlation Between Location & Rating
# ----------------------------------------
# Select required columns
geo_rating_df = df[["Latitude", "Longitude", "Aggregate rating"]].dropna()

# Correlation matrix
correlation = geo_rating_df.corr()
print("\nCorrelation Matrix:")
print(correlation)

# Heatmap visualization
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Between Location and Rating")
plt.show()

```
